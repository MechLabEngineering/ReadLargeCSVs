```python
import pandas as pd
import numpy as np
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

## Timings to read CSV files

MacBook Pro, 2.5GHz Intel Core i5, 8GB 1600MHz DDR3-RAM, SSD

```python
zehnMB={}
zehnMB['CSV'] = 0.196
zehnMB['numpy'] = 1.73
zehnMB['Matlab'] = 1.953
zehnMB['Pandas'] = 0.241
zehnMB['Blaze'] = 0.012
```

```python
hundertMB={}
hundertMB['CSV'] = 1.97
hundertMB['numpy'] = 17.9
hundertMB['Matlab'] = 22.544
hundertMB['Pandas'] = 2.4
hundertMB['Blaze'] = 0.012
```

```python
einGB={}
einGB['CSV'] = 34.4
einGB['numpy'] = 354.0
einGB['Matlab'] = 544.4
einGB['Pandas'] = 25.7
einGB['Blaze'] = 0.012
```

```python
timings = pd.DataFrame.from_dict([zehnMB, hundertMB, einGB])
timings.index = pd.Index(['10MB', '100MB', '1GB'], name='CSV File Size')
timings
```

```python
timings.T.plot(kind='bar', title='Read CSV files with different Python modules', logy=False, ylim=(1e-3, 6e2))
plt.ylabel('read time from SSD in [s]')
```

```python
timings.plot(kind='bar', title='Read CSV files from disk', logy=False, ylim=(1e-3, 1e3))
plt.ylabel('read time from SSD in [s]')
plt.savefig('read-CSV-from-Disc-Blaze-CSV-Pandas-Python.png', dpi=150, bbox_inches='tight')
```

```python
onegbtiming = timings[timings.index=='1GB'].T
```

```python
plt.annotate?
```

```python
onegbtiming.sort('1GB', ascending=False).plot(kind='bar', rot=20, figsize=(8,3))
plt.ylabel('read time from SSD in [s]')
plt.yticks([60*i for i in range(11)])
plt.annotate('9 minutes!', xy=(0.3, 540), xytext=(1, 560), arrowprops={'arrowstyle': 'simple'}, fontsize=14, va='center')
plt.annotate('25 seconds', xy=(3, 30), xytext=(3, 120), arrowprops={'arrowstyle': 'simple'}, fontsize=14, va='center')
plt.savefig('read-1GB-CSV-from-Disc-Blaze-CSV-Pandas-Python-Matlab.png', dpi=150, bbox_inches='tight')
```

## Pandas Speedup

```python
speedup = (onegbtiming.T.Matlab - onegbtiming.T.Pandas) * 100.0 / onegbtiming.T.Matlab
print('%.1f%% speedup from Matlab to Pandas' % speedup)
```

```python
timings.plot(kind='bar', title='Read CSV files from disk', logy=True, ylim=(1e-3, 1e3))
plt.ylabel('read time from SSD in [s]')
plt.savefig('read-CSV-from-Disc-Blaze-CSV-Pandas-Python.png', dpi=150, bbox_inches='tight')
```

```python
timings.plot(logy=False)
plt.legend(loc=2)
plt.title('Read CSV files with different Python modules')
plt.ylabel('read time from SSD in [s]')
```
